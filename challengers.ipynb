```python
## 9. Palindrome Number #easy

class Solution(object):
    def isPalindrome(self, x):
        x_list = list(str(x))
        reversed_list = []
        for i in x_list:
            reversed_list.insert(0,i)
        if x < 0:
            return False
        else:
            x_list_num = int(''.join(map(str,x_list)))
            reversed_list_num = int(''.join(map(str,reversed_list)))       
            return x_list_num == reversed_list_num
             
sol = Solution()

print(sol.isPalindrome(121))
```

```python
#66 Plus One #easy

digits = [1,2,3,9,9,9,9]
class Solution(object):
    def plusOne(self, digits):
        array_num = int(''.join(map(str,digits)))
        array_plus_one = array_num + 1
        array_plus_one = list(map(int, str(array_plus_one)))
        return array_plus_one

sol = Solution()
print(sol.plusOne(digits))
```

```python
# 58 Length of Last Word

class Solution(object):
    def lengthOfLastWord(self, s):
        s = s[::-1]
        i = 0
        length = 0
        while i <= len(s) -1 and s[i] == ' ':
           i +=1
        while 1 <= len(s)-1 and s[i] != ' ':
           length += 1
           i += 1

        if length == 0:

         return 1
        return length
    
solutio = Solution()

print(solutio.lengthOfLastWord("a"))
        
```

```python
#27 remove element

class Solution(object):
    def removeElement(self, nums, val):
        k = 0
        i = 0
        while i < len(nums):
            if nums[i] != val:
                nums[k] = nums[i] 
                k += 1
            i += 1

        return nums

sol = Solution()
print(sol.removeElement([0,1,2,2,3,0,45,2],2))
         
        
```

```python
# teemo on attack 495
def findPoisonedDuration(timeSeries, duration):
    if not timeSeries:
        return 0
    total = 0

    for i in range(len(timeSeries) - 1):
        gap = timeSeries[i+1] - timeSeries[i]
        total += min(gap, duration)

    return total + duration
```

```python
# 657. Robot Return To Origin

class Solution:
    def judgeCircle(self, moves: str) -> bool:
        origin = [0,0]
        for move in moves:
            if move == 'R':
                origin[0] += 1
            elif move == 'L':
                origin[0] -= 1
            elif move == 'U':
                origin[1] += 1
            elif move == 'D':
                origin[1] -= 1

        if origin == [0,0]:
            return True
        else:
            return False
        
solution = Solution()
print(solution.judgeCircle('UDRLL'))


```

```python
# 412 Fizz Buss 

class Solution(object):
    def fizzBuzz(self, n):
        n = n
        output = []
        while n > 0:
            if n % 5 == 0 and n % 3 == 0:
                output.insert(0, "FizzBuzz")
            elif n % 3 == 0:
                output.insert(0, "Fizz")
            elif n % 5 == 0:
                output.insert(0, "Buzz")
            else: 
                output.insert(0, str(n))
            
            n -= 1
        return output


solution = Solution()
print(solution.fizzBuzz(3))
```
