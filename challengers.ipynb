```python
## 9. Palindrome Number 

class Solution(object):
    def isPalindrome(self, x):
        x_list = list(str(x))
        reversed_list = []
        for i in x_list:
            reversed_list.insert(0,i)
        if x < 0:
            return False
        else:
            x_list_num = int(''.join(map(str,x_list)))
            reversed_list_num = int(''.join(map(str,reversed_list)))       
            return x_list_num == reversed_list_num
             
sol = Solution()

print(sol.isPalindrome(121))
```
